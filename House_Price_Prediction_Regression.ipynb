## Modeling
    we are going to build and train four different types of linear regression models which are the OLS model, Ridge regression model, Lasso regression model, Elastic Net regression model. For all the models, we are going to use the pre-built algorithms provided by the scikit-learn package in python. And the process for all the models are the same, first, we define a variable to store the model algorithm,We used Backward Elemination, next, we fit the train set variables into the model, and finally make some predictions in the test set.

#### In this case overfitting are present
 * Traning error is low and Testing error is high

### Creating function for models:-

```python
def modeller(mo):
    model = mo.fit(xtrain,ytrain)
    tr_pred = model.predict(xtrain)
    ts_pred = model.predict(xtest)
    from sklearn.metrics import mean_squared_error
    tr_err = round(mean_squared_error(ytrain,tr_pred),2)
    ts_err = round(mean_squared_error(ytest,ts_pred),2)
    return tr_err,ts_err
```

```python
modeller(lm)
```

# Regularize

```python
from sklearn.linear_model import Ridge,Lasso
```

```python
rr = Ridge(alpha=1)
modeller(rr)
```

```python
q = 0.0
alphas = []
for i in range(0,1500):
    q = round(q + 0.1,2)
    alphas.append(q)
```

```python
tr = []
ts = []
for i in alphas:
    rr = Ridge(alpha=i)
    tr_err,ts_err = modeller(rr)
    tr.append(tr_err)
    ts.append(ts_err)
```

```python
plt.plot(tr)
plt.plot(ts)
```

# Lasso

```python
q = 0.0
alphas = []
for i in range(0,1500):
    q = round(q + 0.1,2)
    alphas.append(q)

tr = []
ts = []
for i in alphas:
    ls = Lasso(alpha=i)
    tr_err,ts_err = modeller(ls)
    tr.append(tr_err)
    ts.append(ts_err)
plt.plot(tr)
plt.plot(ts)
```

## Eastic Net

```python
# ElasticNet model
from sklearn.linear_model import ElasticNet # ElasticNet algorithm

q = 0.0
alphas = []
for i in range(0,1500):
    q = round(q + 0.1,2)
    alphas.append(q)

tr = []
ts = []
for i in alphas:
    en = ElasticNet(alpha=i)
    tr_err,ts_err = modeller(en)
    tr.append(tr_err)
    ts.append(ts_err)
plt.plot(tr)
plt.plot(ts)
```
